```python
import pandas as pd
import numpy as np

# --- Carregar arquivos ---
customers = pd.read_csv("data/raw/customers.csv")
monthly = pd.read_csv("data/processed/monthly_agg.csv", parse_dates=["month"])

# --- Referência: último mês disponível ---
ref_month = monthly['month'].max()
ref_date = pd.to_datetime(ref_month)

# --- RFM: recency, frequency, monetary ---
last_txn = monthly.groupby('customer_id').agg(last_txn_month=('month','max')).reset_index()
last_txn['recency_days'] = (ref_date - pd.to_datetime(last_txn['last_txn_month'])).dt.days

frequency = monthly.groupby('customer_id').agg(frequency=('txn_count','sum')).reset_index()
monetary = monthly.groupby('customer_id').agg(monetary=('txn_amount_sum','sum')).reset_index()

# --- Products count ---
# Ajuste automático: considera apenas colunas existentes
product_cols = ['checking','savings','credit_card','personal_loan','mortgage']
product_cols_exist = [c for c in product_cols if c in customers.columns]

customers['products_count'] = customers[product_cols_exist].sum(axis=1)

# --- Criar features_table ---
feat = customers.merge(last_txn[['customer_id','recency_days']], on='customer_id', how='left')\
                .merge(frequency, on='customer_id', how='left')\
                .merge(monetary, on='customer_id', how='left')

# --- Preencher NA e salvar ---
feat['frequency'] = feat['frequency'].fillna(0)
feat['monetary'] = feat['monetary'].fillna(0)

# --- Churn 90d ---
last_tx = monthly.groupby('customer_id')['month'].max().reset_index()
last_tx['days_since_last_tx'] = (ref_date - pd.to_datetime(last_tx['month'])).dt.days
feat = feat.merge(last_tx[['customer_id','days_since_last_tx']], on='customer_id', how='left')
feat['churn_90d'] = np.where(feat['days_since_last_tx'] > 90, 1, 0)

# --- Salvar CSV final ---
feat.to_csv("data/processed/features_table.csv", index=False)

print("features_table.csv gerado com sucesso!")
feat.head()
```
